```python
d = {"name": "혼자 공부하는 데이터 분석"}
print(d['name'])
```

<혼자 공부하는 데이터 분석>이라는 도서명을 JSON 객체로 나타내면 다음과 같이 중괄호를 사용하고
파이썬의 딕셔너리와 비슷하게 키key와 값value을 콜론(:)으로 연결

도서명에 해당하는 키는 "name"으로 정하고, 키와 값에 문자열을 쓰려면 항상 큰따옴표(" ")로 감싸 주어야 함


```python
import json
```

```python
d_str = json.dumps(d, ensure_ascii= False)
print(d_str)
```

json.dumps()함수를 사용하면 앞에서 만든 딕셔너리 d를 문자열로 변환

- ensure()_ascii매개변수를 False로 지정한 이유: 딕셔너리 d에 한글이 포함되어있기 때문

```python
print(type(d_str))
```

딕셔너리가 문자열로 제대로 바뀐지 확인

```python
d2 = json.loads(d_str)
print(d2['name'])
```

json.loads() 함수
- JSON 문자열을 파이썬 객체로 변환

```python
print(type(d2))
```

문자열이 파이썬 딕셔너리로 바뀌었는지 확인

but 매번 이렇게 하면 번거로우니 json.loads() 함수에 직접 전달

json 은 대괄호 안에 여러 항목을 나열하여 배열을 표현할 수 있음(리스트와 비슷)

```python
d4_str = """
[
  {"name": "혼자 공부하는 데이터 분석", "author": "박해선", "year": 2022},
  {"name": "혼자 공부하는 머신러닝+딥러닝", "author": "박해선", "year":2020}
]
"""
d4 = json.loads(d4_str)
print(d4[0]['name'])
```

데이터프레임으로 바꾸는 또 다른 방법

```python
x_str = """
<book>
    <name> 혼자 공부하는 데이터 분석 </name>
    <author> 박해선 </author>
    <year> 2022 </year>
</book>
"""
```

먼저 xml 문자열을 x_str 변수에 저장

```python
import xml.etree.ElementTree as et
book = et.fromstring(x_str)
```

xml.etree.ElemenTree 모듈의 fromstring()함수를 사용해 x_str 문자열을 파이썬 객체로 변환

```python
print(type(book))
```

type( ) 함수로 fromstring( ) 함수가 어떤 객체를 반환하는지 book 변수의 타입을 확인

```python
print(book.tag)
```

book 객체의 tag 속성을 출력하여 엘리먼트 이름 확인

```python
book_childs = list(book)
print(book_childs)
```

```python
name, author, year = book_childs
print(name.text)
print(author.text)
print(year.text)
```

```python
name = book.findtext('name')
author = book.findtext('author')
year = book.findtext('year')
print(name)
print(author)
print(year)
```

book 객체의 findtext( ) 메서드를 사용하면 해당하는 자식 엘리먼트를 탐색하여 자동으로 텍스트를 반환

```python
x2_str = """
<books>
    <book>
        <name> 혼자 공부하는 데이터 분석 </name>
        <author> 박해선 </author>
        <year> 2022 </year>
    </book>
    <book>
        <name> 혼자 공부하는 머신러닝 + 딥러닝 </name>
        <author> 박해선 </author>
        <year> 2020 </year>
    </book>
</books>
"""
```

```python
books = et.fromstring(x2_str)
print(books.tag)
```

```python
for book in books.findall('book'):
  name = book.findtext('name')
  author = book.findtext('author')
  year = book.findtext('year')
  print(name)
  print(author)
  print(year)
  print()
```
