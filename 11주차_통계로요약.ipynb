평균 구하기

range()는 하나의 숫자를 입력할 경우 0부터 입력된 숫자 직전까지 반복

```python
x = [10, 20, 30]
sum = 0
for i in range(3):
  sum += x[i]
print('평균', sum / len(x))
```

```python
ns_book6['대출건수'].std()
```

판다스의 std() 메서드를 수식으로 구현하려면 샘풀 개수를 n이 아니라 n-1로 나누어야 함
제곱근은 넘파이의 sqrt()함수를 사용하여 계산

```python
import numpy as np
diff = ns_book6['대출건수'] - ns_book6['대출건수'].mean()
np.sqrt(np.sum(diff**2)/(len(ns_book6)-1))
```

unique 함수를 호출해 고유값(values)과 드장 회수 (counts) 배열 획득

```python
values, counts = np.unique(ns_book6['도서명'], return_counts=True)
```

argmax() 함수를 사용하여 counts 배열에서 가장 큰 값의 인덱스를 찾기

```python
max_idx = np.argmax(counts)
```

values 배열에서 해당 인덱스의 값 출력

```python
values[max_idx]
```
