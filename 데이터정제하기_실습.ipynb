```python
import pandas as pd
ns_df = pd.read_csv('ns_202104.csv', low_memory = False)
ns_df.head()
```

```python
ns_book = ns_df.loc[:, '번호':'등록일자']
ns_book.head()
```

원소별 비교를 활용하여 'Unnamed:13'열이 아닌 것을 표시하는 배열

 결과는 넘파이배열(Numpy array}

원소별 비교 연산으로 불리언 배열을 얻었다.

이를 selected_columns변수에 저장하고 판다스 데이터프레임의 loc메서드에 전달하면 True인 열의 행만 선택

```python
selected_columns = ns_df.columns != 'Unnamed: 13'
ns_book = ns_df.loc[:, selected_columns]
ns_book.head()
```

```python
selected_columns = ns_df.columns != '부가기호'
ns_book = ns_df.loc[:, selected_columns]
ns_book.head()
```

 데이터프레임 중간에 있는 '부가기호'열을 제외

 데이터프레임의 메서드를 사용하여 열을 삭제

 판다스에서는 데이터프레임의 행이나 열을 삭제하는 drop()메서드를 제공

drop()메서드로 열을 삭제하려면 첫 번째 매개변수에 삭제하려는 열 이름을 전달하고 axis매개변수를 1로 지정

```python
ns_book = ns_df.drop('Unnamed: 13', axis =1 )
ns_book.head()
```

여러개 삭제하려면 리스트 형태

```python
ns_book = ns_df.drop(['부가기호', 'Unnamed: 13'], axis = 1)
ns_book.head()
```

지금까지는 ns_df데이터프레임에서 열을 삭제한 결과를 ns_book이라는 변수에 저장

drop()메서드에 inplace매개변수를 True로 지정하면

 현재 선택한 데이터프레임을 바로 수정

새로운 변수를 만들어 데이터프레임 수정 결과를 반환하지않고도 ns_book데이터프레임에서 '주제분류번호'열을 삭제!

```python
ns_book.drop('주제분류번호', axis = 1, inplace = True)
ns_book.head()
```

판다스는 비어있는 값을 NaN으로 표시

 drop()메서드와 비슷한 dropna()메서드는 기본적으로 NaN이 하나 이상 포함된 행이나 열을 삭제, axis=1로 지정하여 ns_df데이터프레임에서 NaN이 포함된 열을 삭제

14개의 열이 5개로 확 줄어들었다. 삭제된 열에는 NaN이 적어도 하나 이상있었나 봅니다.

```python
ns_book = ns_df.dropna(axis = 1)
ns_book.head()
```

모든 값이 NaN인 열을 삭제하려면 dropna()메서드에 how매개변수를 'all'로 지정

```python
ns_book = ns_df.dropna(axis = 1, how='all')
ns_book.head()
```

지금까지 열을 삭제하는 방법

 loc메서드로 필요한 열을 선택하려면 슬라이스 연산자나 불리언 배열을 사용할 수 있다.

 drop()메서드로 열을 삭제하려면 axis=1과 삭제하려는 열을 지정! 삭제할 열이 여러개일 때는 열 이름을 리스트로 전달할 수 있다.

또는 dropna()메서드에 axis=1을 지정하면 NaN이 들어있는 열을 삭제할 수 있다.

이번에는 행을 삭제하는 방법

 행을 삭제할 때도 drop()메서드를 사용

  axis매개변수를 0으로 지정하면 행을 삭제할 수 있지만, 기본값이 0이기 때문에 생략해도 괜찮다. 처음 2개행을 삭제

```python
ns_book2 = ns_book.drop([0,1])
ns_book2.head()
```

행을 선택할 때 자주 사용하는 []연산자를 살펴보겠다. []연산자에 열이름 또는 열 이름 리스트를 전달하여 데이터프레임의 열을 선택할 수 있었다.

```python
ns_book2 = ns_book[2:]
ns_book2.head()
```

```python
ns_book2 = ns_book[0:2]
ns_book2.head()
```

슬라이싱 이외에도 불리언 배열을 사용해서 행을 선택

 행을 선택할 때 가장 즐겨 사용하는 방법이다.

  비교 연산자를 활용해 원하는 행은 True로 표시하고 제외할 행은 False로 표시한 불리언 배열을 만들어 사용!

예를 들어 출판사가 '한빛미디어'인 행만 선택

비교 연산자를 사용하여 원본 데이터프레임 ns_df에서 출판사가 '한빛미디어'인 행만 True인 불리언 배열을 만든다.
 그 다음에 이 불리언 배열을[]연산자에 넣어 준다.

```python
selected_rows = ns_df['출판사'] == '한빛미디어'
ns_book2 = ns_book[selected_rows]
ns_book2.head()
#시험
```

loc메서드에 불리언 배열을 사용하여 행을 선택할 수 있다

```python
ns_book2 = ns_book.loc[selected_rows]
ns_book2.head()
```

한가지 예로! 대출건수가 1,000이하인 행을 모두 삭제하고 싶다면 반대로 ns_book['대출건수']>1000조건을 넣어 대출건수가 1,000이 넘는 행을 선택하면 된다!

```python
ns_book2 = ns_book[ns_book['대출건수']>1000]
ns_book2.head()
```

 열을 삭제할 때와 마찬가지로 행을 삭제할 때도 loc메서드, drop()메서드, dropna()메서드를 모두 사용할 수 있지만,
 **불리언 배열을[]연산자에 전달하는 방법을 자주 사용한다는 것



## 중복된 행 찾기

```python
sum(ns_book.duplicated())
```

\sum()함수가 반환한 값이 0인 것을 보니 True로 표시된 행이 하나도 없다
-> 중복된 행이 없다는 뜻

 이처럼 duplicated()메서드는 기본적으로 데이터프레임에 있는 모든 열을 기준으로 중복된 행을 찾는다!

```python
sum(ns_book.duplicated(subset = ['도서명', '저자', 'ISBN']))
```

사실 ns_book데이터프레임의 모든 행은 '번호'열에 고유한 값을 가지고 있기 때문에 중복된 행이 나올 수가 없습니다.

만약 '도서명','저자','ISBN'을 기준으로 중복된 행이 있는지를 찾아보면 어떨까

일부 열을 기준으로 중복된 행을 찾으려면 duplicated()메서드의 **subset매개변수**에 기준열을 나열한다!!

'도서명','저자','ISBN'을 기준으로 찾으면 중복된 행이 2만개가 넘음

'도서명','저자','ISBN'을 기준으로 어떤 데이터가 중복되었는지 확인해보자!
duplicated()메서드에 keep매개변수를 False로 지정하여 중복된 모든 행을 True로 표시합니다. 이 데이터를 ns_book3변수에 저장하여 확인해보자!

```python
dup_rows = ns_book.duplicated(subset=['도서명', '저자','ISBN'], keep = False)
ns_book3 = ns_book[dup_rows]
ns_book3.head()
```

실행 결과를 확인해보니 첫번째 책 '파친코'는 두 권짜리 도서 그래서'권'열에 1과 2로 표시

 두 번째'보건교사 안은영'은 한 권 짜리 도서인데 두 권이 등록

  아마도 인기가 많아 도서관에서 두권을 구입?

정수 타입인 ‘대출건수’ 열은 sum( ) 메서드를 사용
