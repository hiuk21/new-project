```python
# 1. 분기를 연도로 변환
df_all["year"] = (df_all["period"].str[-4:].astype(int) // 100)  # 201512 -> 2015

# 2. 위치별 + 연도별 상가업소번호 집합 만들기
location_year = df_all.groupby(["건물관리번호", "year"])["상가업소번호"].apply(set).reset_index()
location_year.columns = ["건물관리번호", "year", "업소집합"]

# 3. 연도별로 정렬 후 이전 연도와 비교
location_year = location_year.sort_values(["건물관리번호", "year"])

def calc_survival_replacement(group):
    group = group.reset_index(drop=True)
    survival_rates = []
    replacement_rates = []
    
    for i in range(1, len(group)):
        prev = group.loc[i-1, "업소집합"]
        curr = group.loc[i, "업소집합"]
        
        survived = len(prev & curr) / len(prev) if prev else 0  # 생존율
        replaced = len(curr - prev) / len(curr) if curr else 0  # 교체율
        
        survival_rates.append(survived)
        replacement_rates.append(replaced)
    
    group = group.iloc[1:].copy()
    group["생존율"] = survival_rates
    group["교체율"] = replacement_rates
    return group

result = location_year.groupby("건물관리번호", group_keys=False).apply(calc_survival_replacement)
print(result.head(10))
```

```python
# 1. 위치 식별자 만들기 (한 칸 단위)
df_all["위치ID"] = (
    df_all["건물관리번호"].astype(str) + "_" +
    df_all["층정보"].astype(str) + "_" +
    df_all["호정보"].astype(str)
)

# 2. 위치별 + 연도별 상가업소번호 집합
location_year = df_all.groupby(["위치ID", "year"])["상가업소번호"].apply(set).reset_index()
location_year.columns = ["위치ID", "year", "업소집합"]

# 3. 전체 연도 범위 생성 (공실 포함)
all_years = range(df_all["year"].min(), df_all["year"].max() + 1)
all_locations = location_year["위치ID"].unique()

# 모든 위치x연도 조합 만들기 (공실 = NaN으로 채워짐)
full_index = pd.MultiIndex.from_product([all_locations, all_years], names=["위치ID", "year"])
location_year = location_year.set_index(["위치ID", "year"]).reindex(full_index).reset_index()

# 4. 각 위치별로 가게 생존기간 + 교체횟수 + 공실 계산
records = []

for loc_id, group in location_year.groupby("위치ID"):
    group = group.sort_values("year").reset_index(drop=True)
    current_shop = None
    shop_start = None
    tenure = 0
    replacements = 0
    vacant_years = 0

    for _, row in group.iterrows():
        year = row["year"]
        shops = row["업소집합"]

        if pd.isna(shops) or len(shops) == 0:  # 공실
            if current_shop:
                records.append({
                    "위치ID": loc_id,
                    "상가업소번호": current_shop,
                    "입점년도": shop_start,
                    "퇴점년도": year - 1,
                    "생존기간": tenure,
                    "상태": "폐업"
                })
                current_shop = None
                shop_start = None
                tenure = 0
            vacant_years += 1
        else:
            shop = list(shops)[0]  # 대표 상가
            if current_shop is None:  # 신규 입점
                current_shop = shop
                shop_start = year
                tenure = 1
                replacements += 1
            elif shop == current_shop:  # 동일 상가 유지
                tenure += 1
            else:  # 교체
                records.append({
                    "위치ID": loc_id,
                    "상가업소번호": current_shop,
                    "입점년도": shop_start,
                    "퇴점년도": year - 1,
                    "생존기간": tenure,
                    "상태": "교체"
                })
                current_shop = shop
                shop_start = year
                tenure = 1
                replacements += 1

    # 마지막 상가 처리
    if current_shop:
        records.append({
            "위치ID": loc_id,
            "상가업소번호": current_shop,
            "입점년도": shop_start,
            "퇴점년도": year,
            "생존기간": tenure,
            "상태": "생존중"
        })

shop_history = pd.DataFrame(records)
print(shop_history.head(10))
print(shop_history["상태"].value_counts())
```

```python
# 위치별 교체 횟수 계산
replacement_count = shop_history[shop_history["상태"] == "교체"].groupby("위치ID").size().reset_index()
replacement_count.columns = ["위치ID", "교체횟수"]

# 시각화
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rc('font', family='Malgun Gothic')

plt.figure(figsize=(12, 6))
plt.hist(replacement_count["교체횟수"], bins=30, edgecolor='black')
plt.xlabel("교체 횟수")
plt.ylabel("위치 수")
plt.title("위치별 교체 횟수 분포")
plt.grid(axis='y', alpha=0.5)
plt.show()

print(replacement_count["교체횟수"].describe())
```

```python
# 위치별 + 연도별 상가업소번호 (대표 1개)
pivot_df = df_all.groupby(["위치ID", "year"])["상가업소번호"].first().unstack("year")

# 컬럼명 정리
pivot_df.columns = [f"{c}년" for c in pivot_df.columns]

# 공실은 NaN으로 표현됨
print(pivot_df.head())
```

```python
# 연도 컬럼만 추출
year_cols = [c for c in pivot_df.columns]

# 공실율 = 연도별로 NaN인 위치 비율
vacancy_rate = pivot_df.isna().sum() / len(pivot_df) * 100

# 교체율 = 연도별로 이전 연도랑 다른 상가로 바뀐 비율
replacement_rate = {}
for i in range(1, len(year_cols)):
    prev = pivot_df[year_cols[i-1]]
    curr = pivot_df[year_cols[i]]
    # 둘 다 공실 아닌데 상가가 바뀐 경우
    changed = ((prev.notna()) & (curr.notna()) & (prev != curr)).sum()
    total = prev.notna().sum()
    replacement_rate[year_cols[i]] = changed / total * 100

# 생존율 = 교체율의 반대
survival_rate = {k: 100 - v for k, v in replacement_rate.items()}

print("공실율(%):")
print(vacancy_rate)
print("\n교체율(%):")
print(pd.Series(replacement_rate))
print("\n생존율(%):")
print(pd.Series(survival_rate))
```

```python
# 1. 분기별 피벗
pivot_df = df_all.groupby(["위치ID", "period"])["상가업소번호"].first().unstack("period")

# 분기 컬럼 정렬
pivot_df = pivot_df.reindex(sorted(pivot_df.columns), axis=1)

print(pivot_df.shape)
print(pivot_df.columns.tolist())
```

```python
cols = pivot_df.columns.tolist()

vacancy_rates = []
replacement_rates = []
survival_rates = []

for i, col in enumerate(cols):
    # 공실율 = 해당 분기에 NaN인 비율
    vacancy = pivot_df[col].isna().sum() / len(pivot_df) * 100
    vacancy_rates.append(vacancy)
    
    if i == 0:
        replacement_rates.append(None)
        survival_rates.append(None)
    else:
        prev = pivot_df[cols[i-1]]
        curr = pivot_df[cols[i]]
        
        # 이전 분기에 있었던 위치 기준
        had_shop = prev.notna()
        
        # 생존 = 이전이랑 같은 상가
        survived = (had_shop & (prev == curr)).sum()
        # 교체 = 이전이랑 다른 상가 (공실→입점 포함)
        replaced = (had_shop & (curr.notna()) & (prev != curr)).sum()
        
        total = had_shop.sum()
        survival_rates.append(survived / total * 100)
        replacement_rates.append(replaced / total * 100)

# 데이터프레임으로 정리
rate_df = pd.DataFrame({
    "분기": cols,
    "공실율": vacancy_rates,
    "교체율": replacement_rates,
    "생존율": survival_rates
})

# 시각화
fig, axes = plt.subplots(3, 1, figsize=(16, 12))

axes[0].plot(rate_df["분기"], rate_df["공실율"], marker='o', color='gray')
axes[0].set_title("분기별 공실율 (%)")
axes[0].set_ylabel("%")
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(alpha=0.5)

axes[1].plot(rate_df["분기"], rate_df["교체율"], marker='o', color='red')
axes[1].set_title("분기별 교체율 (%)")
axes[1].set_ylabel("%")
axes[1].tick_params(axis='x', rotation=45)
axes[1].grid(alpha=0.5)

axes[2].plot(rate_df["분기"], rate_df["생존율"], marker='o', color='blue')
axes[2].set_title("분기별 생존율 (%)")
axes[2].set_ylabel("%")
axes[2].tick_params(axis='x', rotation=45)
axes[2].grid(alpha=0.5)

plt.tight_layout()
plt.show()

print(rate_df)
```

```python
# 위치별 교체 횟수 계산
replacement_count = pd.DataFrame()
replacement_count["교체횟수"] = (pivot_df != pivot_df.shift(axis=1)).sum(axis=1)

# 분포
print(replacement_count["교체횟수"].value_counts().sort_index())

# 시각화
plt.figure(figsize=(12, 6))
plt.hist(replacement_count["교체횟수"], bins=40, edgecolor='black')
plt.xlabel("교체 횟수")
plt.ylabel("위치 수")
plt.title("위치별 교체 횟수 분포")
plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python
# 위치별 생존 횟수 (이전 분기랑 같은 상가)
survival_count = (pivot_df == pivot_df.shift(axis=1)).sum(axis=1)

# 위치별 공실 횟수 (NaN인 분기 수)
vacancy_count = pivot_df.isna().sum(axis=1)

print("=== 생존 횟수 분포 ===")
print(survival_count.value_counts().sort_index())

print("\n=== 공실 횟수 분포 ===")
print(vacancy_count.value_counts().sort_index())

# 시각화
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

axes[0].hist(replacement_count["교체횟수"], bins=40, edgecolor='black', color='red', alpha=0.7)
axes[0].set_title("위치별 교체 횟수 분포")
axes[0].set_xlabel("교체 횟수")
axes[0].set_ylabel("위치 수")
axes[0].grid(axis='y', alpha=0.5)

axes[1].hist(survival_count, bins=40, edgecolor='black', color='blue', alpha=0.7)
axes[1].set_title("위치별 생존 횟수 분포")
axes[1].set_xlabel("생존 횟수")
axes[1].set_ylabel("위치 수")
axes[1].grid(axis='y', alpha=0.5)

axes[2].hist(vacancy_count, bins=40, edgecolor='black', color='gray', alpha=0.7)
axes[2].set_title("위치별 공실 횟수 분포")
axes[2].set_xlabel("공실 횟수")
axes[2].set_ylabel("위치 수")
axes[2].grid(axis='y', alpha=0.5)

plt.tight_layout()
plt.show()
```
